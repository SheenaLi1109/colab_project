```python
import nltk
from nltk.corpus import PlaintextCorpusReader

corpus_root = '/content/drive/MyDrive/Colab Text'
wordlists = PlaintextCorpusReader(corpus_root, '.*')

books = wordlists.fileids()
print(books)
```

```python
import nltk
testText = wordlists.words(("Emma Watson.txt"))
single = nltk = nltk.text.Text(testText)

findlist = ["personally","gender","can","equually","society"]
single.dispersion_plot(findlist)
```

```python
books = wordlists.fileids()

cfd = nltk.ConditionalFreqDist((fileid, word)
    for fileid in books
    for word in wordlists.words(fileid))
findlist = ['can','could','may','might','must','will','shall','should']
cfd.tabulate(conditions=books, samples=findlist)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
file_text = open('/content/drive/MyDrive/Colab Text/Emma Watson.txt','r').read()
wordcloud = WordCloud(
    background_color="white",
    width=2000,
    height=1000,
    margin=10,
).generate(file_text)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
